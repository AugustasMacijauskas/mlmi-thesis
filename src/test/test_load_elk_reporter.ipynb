```python
from pathlib import Path

import torch


DIR = Path("../elk-reporters/gpt2/imdb/jovial-murdock")

files = [f for f in (DIR / "reporters").iterdir() if f.suffix == ".pt"]
len(files), files[:5]
```

```python
for f in files:
    print(f.name)
    tns = torch.load(f)
    print(tns.keys())
    print(tns["num_classes"])
    print(tns["in_features"])

    for k, v in tns.items():
        if isinstance(v, torch.Tensor):
            print(k, v.shape)
    
    print("-------------\n")
```
